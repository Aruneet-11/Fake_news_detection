Fake news detection Project

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from google.colab import files
import re
import string
```

```python
df_fake=pd.read_csv('/content/Fake.csv')
df_true=pd.read_csv('/content/True.csv')
```

```python
df_fake.shape
df_fake.head(10)
```

```python
df_true.head(10)
```

```python
df_fake['class']=0
df_true['class']=1
```

```python
print(df_fake.shape)
print(df_true.shape)
```

```python
df_fake_manual_testing=df_fake.tail(10)
for i in range(23480,23470,-1):
  df_fake.drop([i], axis=0, inplace=True)
df_true_manual_testing=df_true.tail(10)
for i in range(21416,21106,-1):
  df_true.drop([i], axis=0,inplace=True)

```

```python
df_manual_testing=pd.concat([df_fake_manual_testing,df_true_manual_testing],axis=0)
df_manual_testing.to_csv("manual_testing.csv")
files.download('manual_testing.csv')
```

Merging the datset i.e fake and true data.

```python
df_merge=pd.concat([df_fake,df_true],axis=0)
df_merge.head(20)
```

```python
#removing the title,subject and data column from the datset.
df=df_merge.drop(['title','subject','date'], axis=1)
df.head(10)
```

```python
#to shuffle the data.
df=df.sample(frac=1)
df.head(10)
```

```python
#to check the null values in the dataframe
df.isnull().sum()
```

```python
#to remove all the unnecessary words from the text.
def word_drop(text):
  text=text.lower()
  text=re.sub('\[.*?\]','',text)
  text=re.sub("\\W"," ",text)
  text=re.sub('https?://\S+|www\.\S+' ,' ', text)
  text=re.sub('<.*?>+', ' ',text)
  text=re.sub('[%s]' % re.escape(string.punctuation),' ',text)
  text=re.sub('\n',' ',text)
  text=re.sub('\w*\d\w*',' ',text)
  return text
```

```python
df['text']=df['text'].apply(word_drop)
df.head(10)
```

Define dependent and independent variables.

```python
x=df['text']
y=df['class']
```

```python
#training and testing of the data.
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
Vectorization=TfidfVectorizer()
xv_train=Vectorization.fit_transform(x_train)
xv_test=Vectorization.transform(x_test)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
#check the accuracy of the model
LR.score(xv_test,y_test)
```

```python
#classification report 
pred_LR=LR.predict(xv_test)
print(classification_report(y_test,pred_LR))
```

Making a Predictive System

```python
x_new=xv_test[107]
prediction=LR.predict(x_new)
print(prediction)
if prediction==1:
  print("True news")
else:
  print("Fake news")
```
